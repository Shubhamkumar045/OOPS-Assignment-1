```python
Q1. Explain Class and Object with respect to Object-Oriented Programming. Give a suitable example.
Answer--In object-oriented programming (OOP), a class is a blueprint or template for creating objects.
        It defines the attributes (data) and methods (functions) that the objects of that class will have.
        An object, on the other hand, is an instance of a class.
        It represents a specific instance or occurrence of the concept defined by the class.
Class: A class is like a blueprint or a prototype that defines the structure and behavior of objects.
It encapsulates data and methods that are common to a certain type of entity.

Object: An object is a specific instance of a class.
It's a concrete entity that can hold data and perform actions based on the methods defined in the class.

class Car:
    def __init__(self, make, model, year):
        self.make = make
        self.model = model
        self.year = year
        self.speed = 0

    def accelerate(self):
        self.speed += 10

    def brake(self):
        if self.speed >= 10:
            self.speed -= 10
        else:
            self.speed = 0

    def get_speed(self):
        return self.speed

# Creating instances (objects) of the Car class
car1 = Car("Toyota", "Camry", 2022)
car2 = Car("Honda", "Civic", 2021)

# Using the methods of the Car class on objects
car1.accelerate()
car1.accelerate()
car2.accelerate()

print("Car 1 Speed:", car1.get_speed())  # Output: Car 1 Speed: 20
print("Car 2 Speed:", car2.get_speed())  # Output: Car 2 Speed: 10

car1.brake()
print("Car 1 Speed after braking:", car1.get_speed())  # Output: Car 1 Speed after braking: 10


Q2. Name the four pillars of OOPs.
Answer--1-Encapsulations
        2-Abstraction
        3-Inheritance
        4-Polymorphism
Q3. Explain why the __init__() function is used. Give a suitable example.
Answer--The __init__() function is a special method in Python classes that is used to initialize the attributes of an object when it is created. 
        It's also known as the constructor method. 
        When you create an instance of a class, the __init__() method is automatically called,
        allowing you to set initial values for the object's attributes.
        1--Initiallization
        2--Attribute Assignment
        3--Constructor Logic
class Student:
    def __init__(self, name, age, grade):
        self.name = name
        self.age = age
        self.grade = grade

    def display_info(self):
        print(f"Name: {self.name}, Age: {self.age}, Grade: {self.grade}")

# Creating instances of the Student class
student1 = Student("Alice", 18, "A")
student2 = Student("Bob", 17, "B")

# Displaying information using the display_info method
student1.display_info()
student2.display_info()

Q4. Why self is used in OOPs?
Answer--In object-oriented programming (OOP) in Python, the self keyword is used to refer to the instance of the class itself.
It is a convention and a required parameter in method definitions for classes. 
When you define methods within a class, the first parameter of those methods should be self, which represents the instance of the class 
that the method is being applied to.

Here's why self is used in OOP:

Instance Reference: self allows methods to access and modify the attributes (data) of the instance it belongs to.
It essentially provides a reference to the specific object on which the method is being invoked.

Attribute Access: By using self, you can access the instance's attributes within methods.
It helps to distinguish between instance variables and local variables, allowing you to work with the specific object's data.

Method Invocation: Using self, you can call other methods within the same class, 
allowing methods to collaborate and interact with each other based on the specific instance's data.

Instance Identification: In cases where there are multiple instances of the same class, 
self helps differentiate between them. Each instance has its own set of attributes, and using self ensures that the correct instance's
attributes are being accessed or modified.

Here's an example to illustrate the use of self in OOP:

python
Copy code
class Person:
    def __init__(self, name, age):
        self.name = name
        self.age = age

    def display_info(self):
        print(f"Name: {self.name}, Age: {self.age}")

# Creating instances of the Person class
person1 = Person("Alice", 25)
person2 = Person("Bob", 30)

# Calling the display_info method using self
person1.display_info()  # Output: Name: Alice, Age: 25
person2.display_info()  # Output: Name: Bob, Age: 30
In this example, the self parameter allows the methods (__init__() and display_info()) to access and 
manipulate the attributes of specific instances (person1 and person2). Without self, the methods wouldn't know which
instance's attributes to work with


Q5. What is inheritance? Give an example for each type of inheritance.
Answer--Inheritance is a fundamental concept in object-oriented programming (OOP) that
        allows a new class (subclass or derived class) to inherit attributes and methods from an existing class (base class or parent class)
A--Single Inheritance: In single inheritance, a class inherits from only one base class. This is the simplest form of inheritance.
class Vehicle:
    def display(self):
        print("This is a vehicle.")

class Car(Vehicle):
    def show_info(self):
        print("This is a car.")

car = Car()
car.display()  # Output: This is a vehicle.
car.show_info()  # Output: This is a car.

B0
```
